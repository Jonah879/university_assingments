***a) Number Conversion - Part 1***

Write a corresponding algorithm decimal_int_to_octal() in Python to convert non-negative integer numbers in decimal representation to octal.


```python
def decimal_int_to_octal(n):
    octal = []
    while n>0:
        x = n % 8
        n = n//8
        octal.append(x)
    number = octal.reverse()
    number = "".join(str(x) for x in octal)
    return number

decimal_int_to_octal(200)
```

***b) Number Conversion - Part 2***

Develop a more general algorithm decimal_to_octal() that also converts numbers that include decimal places from decimal representation to octal.

```python
def decimal_to_octal(n):
    # number needs to be above 0
    if n == 0:
        return "0"
# we divide our decimal number into two parts, because each part needs different calculations
# the first part is our integer the second one our float
    first_part = int(n)
    second_part = n - int(n)
# as above, we create an empty list in which we store each number we calculate
    octal = []
    while first_part>0:
        x = first_part % 8 # this is our remainder 
        first_part = first_part//8 # here we are doing a floor division, so the result without the remainder
        octal.append(x) # storing the number in our list
    number = octal.reverse() # reversing it, because we start from the bottom
    number = "".join(str(x) for x in octal) # we dont want a list as our output but a str. converting a list to a string

    octal2 = [] # same as above: creating an empty list in which we store our result
    while second_part>0: 
        second_part *= 8 # multiply the second part with 8
        y = int(second_part) # extracting the integer part from the result, because with the part after the komma we need to calculate again
        octal2.append(y) # adding the integer part to our list, no reverse needed
        second_part -= y # subtracting the integer part to get the decimal part
    number2 = "".join(str(d) for d in octal2)
    
    
    return number + "." + number2
    

decimal_to_octal(55.5)
```
